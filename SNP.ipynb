```python
import yfinance as yf
import pandas as pd
from pytz import timezone

def get_historical_market_cap(ticker, date):
    """
    Fetches the market cap of a stock by calculating it from the close price at 16:00 (market close)
    using yfinance, only considering the data between 9:30 and 16:00 on the specified date.
    """
    try:
        # Set timezone to US/Eastern to align with market hours
        eastern = timezone('US/Eastern')
        
        # Define the start and end times in Eastern Time (NY market time)
        start_time = eastern.localize(pd.Timestamp(f'{date} 09:30:00'))
        end_time = eastern.localize(pd.Timestamp(f'{date} 16:00:00'))
        
        # Fetch historical intraday data (1-minute intervals) for the given date
        stock = yf.Ticker(ticker)
        historical_data = stock.history(start=start_time, end=end_time, interval='1m')

        # Convert index to US/Eastern time to handle the time zone difference
        historical_data.index = historical_data.index.tz_convert('US/Eastern')

        # Filter data between 9:30 AM and 4:00 PM (market hours)
        historical_data = historical_data.between_time('09:30', '16:00')
        
        # Get the close price at market close (16:00)
        if not historical_data.empty:
            close_price = historical_data['Close'].iloc[-1]  # Get the last close price of the day (16:00)
            
            # Fetch current shares outstanding
            shares_outstanding = stock.info.get('sharesOutstanding', 0)
            
            if shares_outstanding:
                market_cap = close_price * shares_outstanding
                return market_cap
            else:
                print(f"Shares outstanding not available for {ticker}")
                return 0
        else:
            print(f"No intraday data available for {ticker} on {date}")
            return 0
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return 0

def filter_bottom_10_percent(watchlist, date):
    """
    Filters out the bottom 10% of stocks by market cap on a specific date between market open (9:30 AM) and close (4:00 PM).
    """
    # Create a list of tuples (ticker, market cap on the specified date)
    market_caps = [(ticker, get_historical_market_cap(ticker, date)) for ticker in watchlist]
    
    # Convert to DataFrame for easier sorting and manipulation
    df = pd.DataFrame(market_caps, columns=['Ticker', 'MarketCap'])
    
    # Sort by MarketCap in ascending order
    df_sorted = df.sort_values(by='MarketCap')
    
    # Calculate the number of stocks to remove (bottom 10%)
    n_to_remove = int(len(df_sorted) * 0.10)
    
    # Remove the bottom 10% of stocks
    df_filtered = df_sorted.iloc[n_to_remove:]
    
    # Return the tickers of the remaining stocks
    return df_filtered['Ticker'].tolist()
watchlist = ['NVDA', 'TSLA', 'INTC', 'AMD', 'F', 'AAPL', 'BAC', 'AMZN', 'WBD', 'AAL', 'T', 'CCL', 'PFE', 'SBUX', 'CMG', 'AVGO', 'GOOGL', 'MU', 'WBA', 'CSCO', 'KEY', 'WMT', 'GOOG', 'MSFT', 'VZ', 'WFC', 'CMCSA', 'KO', 'KVUE', 'OXY', 'HPE', 'SMCI', 'PARA', 'NKE', 'PCG', 'XOM', 'NCLH', 'UBER', 'PYPL', 'META', 'HBAN', 'GM', 'C', 'BMY', 'KMI', 'DIS', 'FCX', 'AMCR', 'CSX', 'CVS', 'LUV', 'SCHW', 'DAL', 'HPQ', 'DG', 'SLB', 'K', 'MO', 'MRK', 'HST', 'HAL', 'CNP', 'QCOM', 'BA', 'AMAT', 'CVX', 'VTRS', 'TFC', 'AES', 'DLTR', 'LVS', 'KHC', 'MDT', 'NEM', 'EQT', 'RF', 'UAL', 'JPM', 'USB', 'CRM', 'MRO', 'NEE', 'CRWD', 'TGT', 'EW', 'BEN', 'APH', 'APA', 'TJX', 'EL', 'JNJ', 'DVN', 'BBWI', 'KDP', 'COP', 'ON', 'V', 'CAG', 'MNST', 'CTRA', 'CFG', 'PG', 'CZR', 'TXN', 'MCHP', 'VST', 'PPL', 'MDLZ', 'WDC', 'BSX', 'DXCM', 'ABNB', 'ORCL', 'WMB', 'RTX', 'EBAY', 'FTNT', 'EXC', 'SW', 'BKR', 'MGM', 'GILD', 'BAX', 'ABT', 'DOC', 'PEP', 'UPS', 'MS', 'KR', 'PANW', 'ABBV', 'GLW', 'O', 'PM', 'MRNA', 'CPRT', 'GE', 'SO', 'IP', 'BBY', 'MMM', 'ALB', 'IVZ', 'VICI', 'MTCH', 'HD', 'TPR', 'ADI', 'WY', 'BX', 'DOW', 'BK', 'CL', 'FITB', 'CNC', 'LLY', 'AIG', 'TMUS', 'NI', 'D', 'GIS', 'APTV', 'ETSY', 'KIM', 'NFLX', 'GEV', 'CARR', 'MOS', 'EMR', 'LULU', 'MCD', 'JCI', 'MET', 'IPG', 'AEP', 'SYF', 'XEL', 'DUK', 'EOG', 'AXP', 'CPB', 'VLO', 'FIS', 'IBM', 'HRL', 'CEG', 'MPC', 'ROST', 'OKE', 'WYNN', 'CTVA', 'JNPR', 'GEN', 'HON', 'PLD', 'FE', 'SYY', 'ETN', 'WEC', 'WELL', 'STX', 'FTV', 'PGR', 'BWA', 'LOW', 'KKR', 'ADM', 'PSX', 'PCAR', 'FOXA', 'ES', 'NWSA', 'ULTA', 'UNH', 'ICE', 'INVH', 'ENPH', 'NRG', 'VTR', 'LW', 'AMT', 'PEG', 'DHI', 'SRE', 'FI', 'AFL', 'RCL', 'CAH', 'CAT', 'NDAQ', 'CTSH', 'UNP', 'ADBE', 'HWM', 'NXPI', 'CDNS', 'IR', 'MA', 'CMS', 'ACN', 'COF', 'NTAP', 'DE', 'STT', 'EXPE', 'YUM', 'DHR', 'FAST', 'GPN', 'TSN', 'FSLR', 'PHM', 'DD', 'BLDR', 'ANET', 'CSGP', 'KMB', 'COST', 'TRGP', 'CCI', 'ALL', 'EVRG', 'TAP', 'A', 'EIX', 'ADSK', 'ZTS', 'INCY', 'PNC', 'LYB', 'GS', 'KEYS', 'OTIS', 'CTLT', 'AMGN', 'CF', 'MKC', 'LKQ', 'HES', 'ETR', 'EA', 'HOLX', 'ED', 'GEHC', 'AKAM', 'UDR', 'HLT', 'LEN', 'BALL', 'INTU', 'TTWO', 'SWKS', 'TER', 'KMX', 'ZBH', 'COO', 'IRM', 'MAR', 'BDX', 'PAYX', 'LYV', 'WM', 'EQR', 'MAS', 'PRU', 'DRI', 'CME', 'NUE', 'ROL', 'ODFL', 'CLX', 'DAY', 'CB', 'SJM', 'HSY', 'DLR', 'STLD', 'SPG', 'SOLV', 'ACGL', 'HSIC', 'FANG', 'AEE', 'MMC', 'WRB', 'AME', 'SYK', 'TEL', 'VLTO', 'CHD', 'SHW', 'TRMB', 'ADP', 'LRCX', 'HIG', 'LNT', 'ISRG', 'GDDY', 'COR', 'IFF', 'JBL', 'DFS', 'CBRE', 'PNR', 'PPG', 'SNPS', 'BG', 'CI', 'LIN', 'SWK', 'CDW', 'OMC', 'CHRW', 'GD', 'NSC', 'NTRS', 'MCK', 'XYL', 'VRTX', 'CBOE', 'TROW', 'HAS', 'QRVO', 'CHTR', 'GL', 'FDX', 'WAB', 'TMO', 'STZ', 'BXP', 'VMC', 'BRO', 'RMD', 'HUM', 'ARE', 'FMC', 'TECH', 'PODD', 'NOW', 'HCA', 'JBHT', 'PAYC', 'REG', 'APD', 'LMT', 'TT', 'RJF', 'TRV', 'PFG', 'IQV', 'BIIB', 'MTB', 'CE', 'DTE', 'SPGI', 'EXPD', 'ALGN', 'ECL', 'AOS', 'CPT', 'DOV', 'DPZ', 'EFX', 'DVA', 'AWK', 'RL', 'KLAC', 'GNRC', 'CRL', 'MAA', 'TXT', 'ITW', 'TSCO', 'ROK', 'PTC', 'AON', 'GPC', 'BR', 'EXR', 'CMI', 'PNW', 'DGX', 'RSG', 'PWR', 'ATO', 'MCO', 'ELV', 'SBAC', 'FOX', 'LHX', 'FRT', 'JKHY', 'PSA', 'LDOS', 'EMN', 'J', 'GRMN', 'NOC', 'AVB', 'AJG', 'UHS', 'PH', 'MSI', 'RVTY', 'VRSN', 'MHK', 'MLM', 'ALLE', 'MOH', 'NWS', 'EPAM', 'VRSK', 'MSCI', 'AXON', 'MPWR', 'IEX', 'CPAY', 'PKG', 'BLK', 'REGN', 'AVY', 'FFIV', 'L', 'AMP', 'IDXX', 'EQIX', 'LH', 'HUBB', 'WST', 'AIZ', 'STE', 'MKTX', 'WTW', 'CINF', 'ROP', 'URI', 'ZBRA', 'ORLY', 'ANSS', 'TFX', 'WAT', 'DECK', 'ESS', 'CTAS', 'NDSN', 'EG', 'POOL', 'BIO', 'HII', 'IT', 'BKNG', 'FDS', 'TDG', 'SNA', 'GWW', 'TYL', 'TDY', 'FICO', 'AZO', 'MTD']
date = '2024-09-05'  # Specify the historical date (YYYY-MM-DD)

# Filter the watchlist
filtered_watchlist = filter_bottom_10_percent(watchlist, date)

# Print the filtered watchlist
print("Filtered Watchlist (Excluding Bottom 10% by Market Cap on {date}):")
print(filtered_watchlist)
```

```python
#monthly volume(4 weeks)
import pandas as pd
import os
import re
from datetime import datetime, timedelta

def get_top_stocks_by_volume_over_period(folder_path, end_date, period_days=28, top_n=500):
    # Dictionary to accumulate total volumes over the period
    volume_dict = {}
    
    # Convert the end_date string to a datetime object
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
    
    # Calculate the start date of the period
    start_date = end_date - timedelta(days=period_days)
    
    # Iterate over all files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            # Extract the date from the file name using regex
            date_match = re.search(r'_(\d{4}-\d{2}-\d{2}).csv$', file_name)
            if date_match:
                file_date_str = date_match.group(1)
                file_date = datetime.strptime(file_date_str, '%Y-%m-%d')
                
                # Only process files within the desired date range
                if start_date <= file_date <= end_date:
                    file_path = os.path.join(folder_path, file_name)
                    
                    # Read the CSV file
                    df = pd.read_csv(file_path)
                    
                    # Check if 'Volume' column exists
                    if 'Volume' not in df.columns:
                        raise ValueError(f"The 'Volume' column is not present in the data for {file_name}.")
                    
                    # Sum up the volume for the stock
                    total_volume = df['Volume'].sum()
                    
                    # Extract the stock ticker from the file name
                    ticker = file_name.split('_')[0]
                    
                    # Accumulate the total volume for the ticker
                    if ticker in volume_dict:
                        volume_dict[ticker] += total_volume
                    else:
                        volume_dict[ticker] = total_volume
    
    # Convert the dictionary to a DataFrame
    volume_df = pd.DataFrame(list(volume_dict.items()), columns=['Ticker', 'Total Volume'])
    
    # Sort the DataFrame by total volume in descending order
    sorted_volume_df = volume_df.sort_values(by='Total Volume', ascending=False)
    
    # Get the top N stocks by total volume
    top_stocks = sorted_volume_df.head(top_n)
    
    # Extract the tickers from the sorted DataFrame and return as a list
    tickers_list = top_stocks['Ticker'].tolist()
    
    return tickers_list

# Example usage
folder_path = r'C:\Users\danym\finance\SNP500'  # Replace with the path to your folder

def list_for_date(end_date):
    #end_date = '2024-10-15'  # Replace with the desired end date in YYYY-MM-DD format
    top_tickers_list = get_top_stocks_by_volume_over_period(folder_path, end_date)

    # Print the list of top tickers
    print(f"Top tickers by total volume over the past 4 weeks ending on {end_date}:")
    print(top_tickers_list)

list_for_date('2024-10-15')
list_for_date('2024-10-16')
list_for_date('2024-10-17')
list_for_date('2024-10-18')
list_for_date('2024-10-21')
list_for_date('2024-10-22')
list_for_date('2024-10-23')
list_for_date('2024-10-24')
list_for_date('2024-10-25')
list_for_date('2024-10-28')
list_for_date('2024-10-29')
list_for_date('2024-10-30')
list_for_date('2024-10-31')

```

```python
#1 minutes volume
import pandas as pd
import os
import re

def has_no_skips(df):
    """
    Check if the DataFrame has minute-by-minute data without skips.
    """
    # Ensure 'Datetime' column exists
    if 'Datetime' not in df.columns:
        return False  # Skip if there is no 'Datetime' column
    
    # Parse the 'Datetime' column to datetime objects
    df['Datetime'] = pd.to_datetime(df['Datetime'])
    
    # Set 'Datetime' as the index
    df = df.set_index('Datetime')
    
    # Create the expected time range with minute frequency
    expected_freq = pd.date_range(start=df.index.min(), end=df.index.max(), freq='T')
    
    # Get the actual timestamps in the dataframe
    actual_freq = df.index
    
    # Check if all expected minutes are present in the actual data
    has_skips = len(expected_freq) != len(actual_freq) or not all(expected_freq == actual_freq)
    
    return not has_skips

def get_top_stocks_by_daily_volume(folder_path, date, top_n=500):
    # Dictionary to accumulate total volumes
    volume_dict = {}
    
    # Regex to match the date in the filename
    date_pattern = re.compile(rf'_{date}.csv$')
    
    # Iterate over all CSV files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv') and date_pattern.search(file_name):
            file_path = os.path.join(folder_path, file_name)
            
            # Read the CSV file
            df = pd.read_csv(file_path)
            
            # Check if 'Volume' column exists
            if 'Volume' not in df.columns:
                raise ValueError(f"The 'Volume' column is not present in {file_name}.")
            
            # Check if the data has a 'Datetime' column and minute-by-minute timestamps without skips
            if not has_no_skips(df):
                #print(f"Skipping {file_name} due to missing or incomplete 'Datetime' data.")
                continue
            
            # Sum up the volume for the stock
            total_volume = df['Volume'].sum()
            
            # Extract the stock ticker from the file name
            ticker = file_name.split('_')[0]
            
            # Add the total volume to the dictionary
            volume_dict[ticker] = total_volume
    
    # Convert the dictionary to a DataFrame
    volume_df = pd.DataFrame(list(volume_dict.items()), columns=['Ticker', 'Total Volume'])
    
    # Remove bottom 10% of stocks by total volume
    if not volume_df.empty:
        # Calculate the 10th percentile
        volume_threshold = volume_df['Total Volume'].quantile(0.10)
        
        # Filter out stocks below the 10th percentile
        volume_df = volume_df[volume_df['Total Volume'] > volume_threshold]
    
    # Sort the DataFrame by total volume in descending order
    sorted_volume_df = volume_df.sort_values(by='Total Volume', ascending=False)
    
    # Get the top N stocks by total volume
    top_stocks = sorted_volume_df.head(top_n)
    
    # Extract the tickers from the sorted DataFrame and return as a list
    tickers_list = top_stocks['Ticker'].tolist()
    
    # Calculate the volume range (min and max) after filtering
    if not volume_df.empty:
        volume_range = (volume_df['Total Volume'].min(), volume_df['Total Volume'].max())
        print(f"\nVolume range after filtering: {volume_range[0]} (min) - {volume_range[1]} (max)")
    
    return tickers_list

# Example usage
folder_path = r'C:\Users\danym\finance\SNP500'  # Replace with the path to your folder
date = '2024-09-04'  # Replace with the desired date in YYYY-MM-DD format
top_tickers_list = get_top_stocks_by_daily_volume(folder_path, date)

# Print the list of top tickers
print(f"\nTop tickers by total daily volume on {date}:")
print(top_tickers_list)
print(len(top_tickers_list))
```
